# clean, stem and tokenize cliente_tabla.csv

```python
# load cliente_tabla.csv
cliente_tabla = pd.read_csv("./input/cliente_tabla.csv")
cliente_tabla
```

```python
# delete duplicated rows of the table
cliente_tabla=cliente_tabla.drop_duplicates('Cliente_ID',keep='first')
cliente_tabla
```

# Stemming

```python
# remove numbers from the string
punctuation = re.compile(r'[-.?!,":;()|0-9]')
cliente_tabla.ix[:,'NombreCliente']= [punctuation.sub("",word) for word in cliente_tabla.ix[:,'NombreCliente']]
cliente_tabla
```

# stem NombreClient

```python
stemmer = nltk.SnowballStemmer('spanish')
# for idx in cliente_tabla.index.values:
#     cliente_tabla.ix[idx,'NombreCliente'] = stemmer.stem(unicode(cliente_tabla.ix[idx,'NombreCliente'],'utf-8'))
#     if ((idx%1000)==0):
#         print("The %dth obs" % idx)
cliente_tabla.ix[:,'NombreCliente'] = [stemmer.stem(unicode(words,'utf-8')) for words in cliente_tabla.ix[:,'NombreCliente']]
cliente_tabla
```

```python
# tokenize cliente names
cliente_tabla.ix[:,'NombreSplitStem'] = cliente_tabla.ix[:,'NombreCliente'].str.split(' ')
cliente_tabla
```

```python
cliente_tabla.to_csv('./input/cliente_tabla_clean.csv',columns=['Cliente_ID','NombreCliente','NombreSplitStem'], index=False,encoding='utf-8')
```

# stem and split town_state.csv

```python
# load producto_tabla.csv
town_state = pd.read_csv("./input/town_state.csv")
town_state
```

```python
# delete duplicated rows of the table
town_state=town_state.drop_duplicates('Agencia_ID',keep='first')
town_state
```

```python
# clean puctuations
punctuation = re.compile(r'[-.?!,":;()]')
town_state.ix[:,'TownStem']= [punctuation.sub("",word) for word in town_state.ix[:,'Town']]
town_state.ix[:,'StateStem']= [punctuation.sub("",word) for word in town_state.ix[:,'State']]
town_state
```

```python
stemmer = nltk.SnowballStemmer('spanish')
# for idx in Town_State.index.values:
#     Town_State.ix[idx,'StateStem'] = stemmer.stem(unicode(Town_State.ix[idx,'StateStem'],'utf-8'))
#     Town_State.ix[idx,'TownStem'] = stemmer.stem(unicode(Town_State.ix[idx,'TownStem'],'utf-8'))
#     if ((idx%500)==0):
#         print("The %dth obs" % idx)
town_state.ix[:,'TownStem'] = [stemmer.stem(unicode(words,'utf-8')) for words in town_state.ix[:,'TownStem']]
town_state.ix[:,'StateStem'] = [stemmer.stem(unicode(words,'utf-8')) for words in town_state.ix[:,'StateStem']]
town_state
```

```python
# tokenize towns and states
town_state.ix[:,'TownStemSplit'] = town_state.ix[:,'TownStem'].str.split(' ')
town_state.ix[:,'StateStemSplit'] = town_state.ix[:,'StateStem'].str.split(' ')
town_state
```

```python
town_state.to_csv('./input/town_state_clean.csv',columns=['Agencia_ID','Town','State','TownStem',
                                                             'StateStem','TownStemSplit','StateStemSplit'],
                     index=False,encoding='utf-8')
```
